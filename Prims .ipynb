```python
import sys

def printMST(parent, graph):
    print("Edge \tWeight")
    for i in range(1, len(graph)):
        print(parent[i], "-", i, "\t", graph[i][parent[i]])
        
    result=0
    for i in range(1,len(graph)):
        result=result + graph[i][parent[i]]
    print("Weight of MST : ",result)

def minKey(key, mstSet):
    min_val = sys.maxsize
    min_index = -1
    for v in range(len(key)):
        if key[v] < min_val and not mstSet[v]:
            min_val = key[v]
            min_index = v
    return min_index

def primMST(graph):
    V = len(graph)
    key = [sys.maxsize] * V
    parent = [None] * V
    key[0] = 0
    mstSet = [False] * V
    parent[0] = -1

    for _ in range(V):
        u = minKey(key, mstSet)
        mstSet[u] = True
        for v in range(V):
            if graph[u][v] > 0 and not mstSet[v] and key[v] > graph[u][v]:
                key[v] = graph[u][v]
                parent[v] = u

    printMST(parent, graph)

def get_graph_input(nodes, edges):
    matrix = [[0 for _ in range(nodes)] for _ in range(nodes)]
    for i in range(edges):
        start = int(input(f"Enter Start Vertex of Edge {i}: "))
        end = int(input(f"Enter End Vertex of Edge {i}: "))
        weight = int(input(f"Enter Weight for Edge {i}: "))
        matrix[start][end] = weight
        matrix[end][start] = weight
    return matrix

# Driver's code
nodes = int(input("Enter Total Number of Nodes: "))
edges = int(input("Enter Total Number of Edges: "))
    
graph = get_graph_input(nodes, edges)
primMST(graph)
```
