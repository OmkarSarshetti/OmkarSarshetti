```python
import sys

def printSolution(dist):
    print("Vertex \tDistance from Source")
    for node in range(len(dist)):
        print(node, "\t", dist[node])

def minDistance(dist, sptSet):
    min = sys.maxsize
    for u in range(len(dist)):
        if dist[u] < min and not sptSet[u]:
            min = dist[u]
            min_index = u
    return min_index

def dijkstra(graph, src):
    V = len(graph)
    dist = [sys.maxsize] * V
    dist[src] = 0
    sptSet = [False] * V

    for _ in range(V):
        x = minDistance(dist, sptSet)
        sptSet[x] = True

        for y in range(V):
            if graph[x][y] > 0 and not sptSet[y] and dist[y] > dist[x] + graph[x][y]:
                dist[y] = dist[x] + graph[x][y]

    printSolution(dist)

nodes = int(input("Enter Total Number of Nodes: "))
edges = int(input("Enter Total Number of Edges: "))
matrix = [[0 for _ in range(nodes)] for _ in range(nodes)]

for i in range(edges):
    start = int(input(f"Enter Start Vertex of Edge {i+1}: "))
    end = int(input(f"Enter End Vertex of Edge {i+1}: "))
    weight = int(input(f"Enter Weight for Edge {i+1}: "))
    matrix[start][end] = weight
    matrix[end][start] = weight

source = int(input("Enter the source vertex: "))
dijkstra(matrix, source)
```
